```python
#8. How do you count how many unique rows a DataFrame has (i.e. ignore all rows that are duplicates)?

dictDup={'ColA':[1,1,1,1,1,2,3],
     'ColB': [1,1,1,2,2,1,2]}
     
dfDup=pd.DataFrame(dictDup)
print(dfDup)
len(dfDup.groupby(['ColA','ColB']))
print (dfDup.groupby(['ColA','ColB']).size().reset_index(name='Count'))
```

```python
# 9. How to keep only top 2 most frequent values as it is and replace everything else as ‘Other’? Take the following as input.

np.random.RandomState(100)
ser = pd.Series(np.random.randint(1, 5, [12]))

print("Original:")
print(ser)

print("Freq: \n", ser.value_counts())

result = ser[~ser.isin(ser.value_counts().index[:2])] = 'Other'
print(ser)
```
